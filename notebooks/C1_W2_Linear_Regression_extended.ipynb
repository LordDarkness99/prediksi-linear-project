```python
import matplotlib.pyplot as plt
import pandas as pd
import joblib

# Load data
df = pd.read_csv('../data/sample_salary.csv')

# Load model
model = joblib.load('../models/manual_model.joblib')

# Prediksi
X = df['experience'].values.reshape(-1, 1)
y = df['salary']
y_pred = model.predict(X)

# Plot
plt.scatter(X, y, label='Data Asli')
plt.plot(X, y_pred, color='red', label='Prediksi')
plt.xlabel('Pengalaman Kerja (tahun)')
plt.ylabel('Gaji (Rupiah)')
plt.title('Prediksi Gaji Berdasarkan Pengalaman Kerja')
plt.legend()
plt.show()
```
